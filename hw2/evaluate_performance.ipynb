```python
import os
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
```

```python
def calculate_scores(label_path, path):

    test_answer = pd.read_csv(label_path)
    number_of_test = len(test_answer)
    pred_test = pd.read_csv(path)

    if len(pred_test)!=number_of_test:
        print('The number of predicted labels does not match the number of ground truth labels.')
        print('Ground-truth:', number_of_test, 'Predicted:', len(pred_test))
        return 0, 0, len(pred_test)

    test_answer.set_index(['dialog_id', 'speaker', 'start_time', 'end_time'], inplace=True)
    pred_test.set_index(['dialog_id', 'speaker', 'start_time', 'end_time'], inplace=True)
    merged_df = test_answer.merge(pred_test, left_index=True, right_index=True, suffixes=('_df1', '_df2'))
    acc = accuracy_score(merged_df['da_tag_df1'], merged_df['da_tag_df2'])
    f1 = f1_score(merged_df['da_tag_df1'], merged_df['da_tag_df2'], average='macro')

    return acc, f1, len(pred_test)
```

```python
# path to the .csv that includes your predictions
predicted_label_path = 'valid_ab1234_speech.csv'
# path to the .csv that has ground-truth labels
gt_label_path = "valid.csv"

acc, f1, length = calculate_scores(gt_label_path, predicted_label_path)

print("Accuracy:", acc)
print("F1:", f1)
```
